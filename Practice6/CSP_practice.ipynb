```python
import time
import numpy as np

N = 16


def my_range(start, end):
    if start <= end:
        return range(start, end + 1)
    else:
        return range(start, end - 1, -1)


class Problem:
    char_mapping = ('·', 'Q')

    def __init__(self, n=4):
        self.n = n

    def is_valid(self, state):
        """
        check the state satisfy condition or not.
        :param state: list of points (in (row id, col id) tuple form)
        :return: bool value of valid or not
        """
        board = self.get_board(state)
        res = True
        for point in state:
            i, j = point
            condition1 = board[:, j].sum() <= 1
            condition2 = board[i, :].sum() <= 1
            condition3 = self.pos_slant_condition(board, point)
            condition4 = self.neg_slant_condition(board, point)
            res = res and condition1 and condition2 and condition3 and condition4
            if not res:
                break
        return res

    def is_satisfy(self, state):
        return self.is_valid(state) and len(state) == self.n

    def pos_slant_condition(self, board, point):
        i, j = point
        tmp = min(self.n - i - 1, j)
        start = (i + tmp, j - tmp)
        tmp = min(i, self.n - j - 1)
        end = (i - tmp, j + tmp)
        rows = my_range(start[0], end[0])
        cols = my_range(start[1], end[1])
        return board[rows, cols].sum() <= 1

    def neg_slant_condition(self, board, point):
        i, j = point
        tmp = min(i, j)
        start = (i - tmp, j - tmp)
        tmp = min(self.n - i - 1, self.n - j - 1)
        end = (i + tmp, j + tmp)
        rows = my_range(start[0], end[0])
        cols = my_range(start[1], end[1])
        return board[rows, cols].sum() <= 1

    def get_board(self, state):
        board = np.zeros([self.n, self.n], dtype=int)
        for point in state:
            board[point] = 1
        return board

    def print_state(self, state):
        board = self.get_board(state)
        print('_' * (2 * self.n + 1))
        for row in board:
            for item in row:
                print(f'|{Problem.char_mapping[item]}', end='')
            print('|')
        print('-' * (2 * self.n + 1))
```

# BTS: Backtracking search (Request 1, DDL: 10.28)

```python
def bts(problem: Problem):
    action_stack = []
    temp = []
    record = []
    row = 0
    while not problem.is_satisfy(action_stack):
        i = 0
        while i < N:
            pos = (row, i)
            temp.append(pos)
            if problem.is_valid(temp):
                action_stack = temp
                record.append(pos)
                row += 1
                break
            else:
                temp.remove(pos)
            if i == N - 1:
                row -= 1
                if record[row][1] < N - 1:
                    i = record[row][1]
                    while len(record) > row:
                        action_stack.remove(record[len(record) - 1])
                        record.remove(record[len(record) - 1])
                else:
                    row -= 1
                    i = record[row][1]
                    while len(record) > row:
                        action_stack.remove(record[len(record) - 1])
                        record.remove(record[len(record) - 1])
            i += 1
        yield action_stack
```

# Improving BTS To DO (Request 2, DDL: 11.04)
* Which variable should be assigned next?
* In what order should its values be tried?
* Can we detect inevitable failure early?

### Minimum Remaining Value
Choose the variable with the fewest legal values in its domain
### Least constraining value
Given a variable, choose the least constraining value: the one that rules out the fewest values in the remaining variables
### Forward checking
Keep track of remaining legal values for the unassigned variables. Terminate when any variable has no legal values.

```python
def advanced_bts(problem: Problem):
    cnt_list = [0 for _ in range(problem.n)]
    chessboard = np.zeros((problem.n, problem.n))

    def change_board(pos):
        r, c = pos[0], pos[1]
        for i in range(problem.n):
            chessboard[r][i] += 1
            chessboard[i][c] += 1
        i = 0
        while i + r < problem.n and i + c < problem.n:
            chessboard[r + i][c + i] += 1
            i += 1
        i = 0
        while i + r < problem.n and c - i >= 0:
            chessboard[r + i][c - i] += 1
            i += 1
        i = 0
        while r - i >= 0 and c - i >= 0:
            chessboard[r - i][c - i] += 1
            i += 1
        i = 0
        while r - i >= 0 and c + i < problem.n:
            chessboard[r - i][c + i] += 1
            i += 1
        chessboard[r][c] -= 5

    def delete_board(pos):
        r, c = pos[0], pos[1]
        for i in range(problem.n):
            chessboard[r][i] -= 1
            chessboard[i][c] -= 1
        i = 0
        while i + r < problem.n and i + c < problem.n:
            chessboard[r + i][c + i] -= 1
            i += 1
        i = 0
        while i + r < problem.n and c - i >= 0:
            chessboard[r + i][c - i] -= 1
            i += 1
        i = 0
        while r - i >= 0 and c - i >= 0:
            chessboard[r - i][c - i] -= 1
            i += 1
        i = 0
        while r - i >= 0 and c + i < problem.n:
            chessboard[r - i][c + i] -= 1
            i += 1
        chessboard[r][c] += 5

    def find_idx():
        for i in range(problem.n):
            cnt = 0
            for k in range(problem.n):
                if chessboard[k][i] == 0:
                    cnt += 1
            cnt_list[i] = cnt
        nonzero_list = [x for x in cnt_list if x != 0]
        if len(nonzero_list) == 0:
            return -1
        min_val = min(nonzero_list)
        return cnt_list.index(min_val)

    action_stack = []
    cur_range = np.array([True] * problem.n * problem.n)
    cur_range = cur_range.reshape(problem.n, problem.n)
    while not problem.is_satisfy(action_stack):
        flag = False
        idx = find_idx()
        if idx != -1:
            for i in range(problem.n):
                if cur_range[i][idx]:
                    pos = (i, idx)
                    action_stack.append(pos)
                    if problem.is_valid(action_stack):
                        flag = True
                        change_board(pos)
                        break
                    else:
                        cur_range[i][idx] = False
                        action_stack.pop()
            if not flag:
                for i in range(problem.n):
                    cur_range[i][idx] = True
        if not flag:
            delete_board(action_stack[len(action_stack) - 1])
            cur_range[action_stack[len(action_stack) - 1][0]][action_stack[len(action_stack) - 1][1]] = False
            action_stack.pop()
        yield action_stack
```
